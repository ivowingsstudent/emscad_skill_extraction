```python
import pandas as pd
import spacy
import csv
nlp = spacy.load("en_core_web_sm")
from tqdm import tqdm
tqdm.pandas()
```

```python
path = "/Users/ivowings/Sync/Thesis/Datasources/Skills/soft_skills_fromsoftskillmatching.csv"
path2 = "/Users/ivowings/Sync/Thesis/Datasources/Skills/soft_skills_fromsoftskillmatching(2).csv"
```

```python
import csv
with open(path, newline='', encoding = 'utf8') as f:
    csvread = csv.reader(f)
    batch_data = list(csvread)

with open(path2, newline='', encoding = 'utf8') as f:
    csvread = csv.reader(f)
    batch_data2 = list(csvread)
```

```python
df1 = pd.DataFrame(batch_data, columns=['is_soft_skill', 'soft_skill', 'context', 'test'])
df1['test'] = df1['test'].fillna('')
df1['context'] = df1['context'] + ' ' + df1['test']
df1 = df1[['is_soft_skill', 'soft_skill', 'context']]
df1 = df1.drop(index=0)
```

```python
df2 = pd.DataFrame(batch_data2, columns=['is_soft_skill', 'soft_skill', 'context'])
df2 = df2.drop(index=0)
```

```python
df = pd.concat([df1,df2]).reset_index()
df
```

```python
def character_counter(text):
    from collections import Counter
    my_str = text
    counter = Counter(my_str)
    return counter['*']
```

```python
df['test'] = df['context'].str.replace(r'[^A-Za-z0-9 ]+', ' ')
df['test'] = df['test'].str.replace(r'xxx ', 'XXX', regex=True)
df['test'] = df['test'].str.replace(r'xxx', 'XXX', regex=True)
df['test'] = df['test'].str.replace(r'(X+)', '* ', regex=True)
df['test'] = df['test'].str.strip()
df['test'] = df['test'].str.lower()
df['test'] = df['test'].str.replace(r'[ ]{2,}', ' ', regex=True)

df['test2'] = df['test'].apply(character_counter)
df

#Removing rows which have multiple soft skills 
df = df[df.test2==1]

```

```python
test = df['test'].str.split('*',expand=True)
df = df.join(test)
df
```

```python
testdf = df[[0,'soft_skill',1, 'is_soft_skill']]
testdf = testdf.rename(columns={0:'left_context', 1:'right_context'})

testdf = testdf.rename(columns={'soft_skill' : 'candidate_skill', 'is_soft_skill':'label'})
testdf
```

```python
#Removing candidate skills with more than 4 tokens in order to standardize the data with the manually labeled data

testdf['length'] = testdf['candidate_skill'].apply(nlp)
testdf['length'] = testdf['length'].apply(len)
testdf = testdf[testdf.length<=4]
testdf
```

```python
testdf.to_csv("/Users/ivowings/Sync/Thesis/Datasources/Skills/Pre annotated soft skills/annotated_soft_skills_context.csv",
         quoting=csv.QUOTE_NONNUMERIC,
                index=False)
```
