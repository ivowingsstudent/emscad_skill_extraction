```python
import pandas as pd
from flashtext import KeywordProcessor
import spacy
import sys
from tqdm import tqdm
import re
pd.set_option("display.max_rows", None, "display.max_columns", None)
nlp = spacy.load("en_core_web_lg")
```

```python
#Function to remove stopwords
def stop_word_remover(text):
    doc = nlp(str(text))
    text = ' '.join([token.text for token in doc if token.is_stop == False])
    return text

#Function to lemmatize
def lemmatizer(text):
    doc = nlp(str(text))
    text = ' '.join([token.lemma_ if token.lemma_ != '-PRON-' else token.text for token in doc])
    return text
```

```python
df3 = pd.read_excel('/Users/ivowings/Sync/Thesis/Datasources/Skills/soft_skills_buildingasoftskilltaxonomy.xlsx')
```

```python
df3 = df3['Soft skill, Alternative labels'].str.split(' ', 1, expand=True)
df3 = df3.rename(columns={0: "soft_skill", 1: "alternative_labels"})
df3['soft_skill'] = df3['soft_skill'].str.replace(',', '')
```

```python
#Only selecting the alternative labels
df3 = pd.DataFrame(df3['alternative_labels'])
```

```python
#Creating the softskills list
df3 = df3['alternative_labels'].str.split(',', expand=True).values.flatten()
df3 = pd.DataFrame({'alternative_labels':df3})
df3 = df3[df3.alternative_labels.isna()==False].reset_index()
#Lowercasing all skills
df3 = pd.DataFrame(df3['alternative_labels'].str.lower())
```

```python
#Combining all dfs
softskills = pd.concat([df3])
```

```python
#Removing trailing whitespaces and any special characters
softskills = softskills.rename(columns={'alternative_labels' : 'soft_skill'})
softskills['soft_skill'] = softskills['soft_skill'].str.replace('[^A-Za-z0-9 ]+', ' ')
softskills['soft_skill'] = softskills['soft_skill'].str.strip()
```

```python
tqdm.pandas()
softskills['soft_skill_lemmatized'] = softskills['soft_skill'].progress_apply(lemmatizer)
softskills['soft_skill_no_stopwords'] = softskills['soft_skill'].progress_apply(stop_word_remover)

softskills.drop_duplicates(subset=None, keep="first", inplace=True)
softskills = softskills.reset_index(drop=True)
```

```python
softskills.to_csv("/Users/ivowings/Sync/Thesis/Datasources/Preprocessed/Skills/softskills.csv", index=False)
```
